```python
text_file = open("Shrek Script.txt")
file_contents = text_file.read()
print(file_contents)
text_file.close()
```

```python
def calculate_frequencies(file_contents):
    # Here is a list of punctuations and uninteresting words you can use to process your text
    punctuations = '''!()-[]{};:'"\,<>./?@#$%^&*_~1234567890'''
    uninteresting_words = ["the", "a", "to", "if", "is", "it", "of", "and", "or", "an", "as", "i", "me", "my", \
    "we", "our", "ours", "you", "your", "yours", "he", "she", "him", "his", "her", "hers", "its", "they", "them", \
    "their", "what", "which", "who", "whom", "this", "that", "am", "are", "was", "were", "be", "been", "being", \
    "have", "has", "had", "do", "does", "did", "but", "at", "by", "with", "from", "here", "when", "where", "how", \
    "all", "any", "both", "each", "few", "more", "some", "such", "no", "nor", "too", "very", "can", "will", "just","huh","\n",\
                           "\r","I","You","Im","on","oh","for","not","Oh","in","know","No"]
    
    # LEARNER CODE START HERE
    file_contents_new = file_contents.strip()
    file_contents_new = list(file_contents_new)
    punctuations_list =[]
    for i in punctuations :
        punctuations_list.append(i)
    half_clean_file_contents = [i for i in file_contents_new if i not in punctuations_list]
    join_half_clean_file_contents = "".join(half_clean_file_contents)
    split_half_clean_file_contents = join_half_clean_file_contents.split()
    clean_file_contents = [i for i in split_half_clean_file_contents if i not in uninteresting_words]
    
    counts = dict()
    for word in clean_file_contents:
        if word in counts:
            counts[word] += 1.0
        else:
            counts[word] = 1.0

    #wordcloud
    cloud = wordcloud.WordCloud()
    cloud.generate_from_frequencies(counts)
    return cloud.to_array()
```
